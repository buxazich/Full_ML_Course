```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('Advertising.csv')
df.head()
```

```python
df['Total_spend'] = df['TV'] + df['radio'] + df['newspaper']
df.head()
```

```python
sns.scatterplot(df, x='Total_spend', y='sales')
```

```python
sns.regplot(df, x='Total_spend', y='sales')
```

```python
X = df['Total_spend']
y = df['sales']
```

```python
# y = mx + b 
# y = B1*x + B0
help(np.polyfit)
```

```python
# B1, B0 
np.polyfit(X,y, deg=1)
```

```python
potential_spend = np.linspace(0,500, 100)
```

```python
predicted_sales = 0.04868788 * potential_spend + 4.24302822
#   
sns.scatterplot(df, x='Total_spend', y='sales')
#   
plt.plot(potential_spend, predicted_sales, color='red');
```

```python
spend = 200 
predicted_sale = 0.04868788 * spend + 4.24302822
predicted_sale
```

```python
# y = B1*x + B0 

np.polyfit(X, y, deg=3)
# y = B3 * x^3 + B2 * x^2 + B1 * x + B0
```

```python
pot_spend = np.linspace(0,500, 100)
prediction = 3.07615033e-07 * pot_spend**3 + \
            -1.89392449e-04 * pot_spend**2 + \
             8.20886302e-02 * pot_spend + 2.70495053e+00

#   
sns.scatterplot(df, x='Total_spend', y='sales')
#   
plt.plot(pot_spend, prediction, color='red');
```
