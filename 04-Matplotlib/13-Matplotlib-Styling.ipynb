# Стилизация Matplotlib 
* Есть разнообразные возможности по стилизицаии и редактировании графиков. 

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(0,11,10)
x
```

```python
fig = plt.figure()

axes_1 = fig.add_axes([0,0,1,1])

axes_1.plot(x,x, label='X vs X')
axes_1.plot(x,x**2, label='X vs X^2')

axes_1.legend(loc='best')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, color='red', label='X VS X', lw='4', linestyle='--') #or RGB code
ax.plot(x,x+1, color='green', label='X+1 VS X+1', marker='o')

# Возможные варианты для linestype:  ‘--‘, ‘–’, ‘-.’, ‘:’, ‘steps’


ax.legend();
```
