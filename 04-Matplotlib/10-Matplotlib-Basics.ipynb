```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
x = [2,23,2,3,5,6,7,4]
y = [10,33,2,3,3,1,5,4]
plt.plot(x,y)
```

```python
x = np.arange(0,10)
y = x ** 2 
plt.plot(x,y)
plt.title('Non-linear')
plt.xlabel('x')
plt.xlim(0,9)
plt.ylabel('y')
plt.ylim(0,15)
plt.savefig('Non-linear.png');
```
