# Объект Matplotlib Figure 
* Объектно-ориентированный API Matplotlib использует объект Figure.
* Далее к объекту Figure добавляются оси (axis).
* И на осях уже рисуются графики 
* Такой подход позволяет контролировать каждый шаг построения графика. 

##### Замечание. 
* объект Figure невидим до тех пор, пока на него не будут добавлены оси. 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
a = np.linspace(0,10,11)
b = a**4
print(a)
print(b)
```

```python
x = np.arange(0,10)
y = 2*x 
print(x)
print(y)
```

```python
fig = plt.figure()
# набор больших осей
axes_1 = fig.add_axes([0,0,1,1]) 
axes_1.plot(a,b)
axes_1.set_xlim(0,9)
axes_1.set_ylim(0,8000)
axes_1.set_xlabel('A')
axes_1.set_ylabel('B')
axes_1.set_title('Возьведение в степень 4')
# набор малых осей
axes_2 = fig.add_axes([0.15,0.35,0.4,0.4])
axes_2.plot(a,b)
axes_2.set_xlim(1,2)
axes_2.set_ylim(0,50)
axes_2.set_xlabel('A')
axes_2.set_ylabel('B')
axes_2.set_title('Увеличенный фрагмент')
;
```

```python
fig = plt.figure(figsize=(5,4), dpi=100)
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b)
fig.savefig('OOP_figure.png')
```
