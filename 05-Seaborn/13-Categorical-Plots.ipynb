# Categorial plots - распределение значении. 

Распределение по категориям:
1. Boxplot.
2. Violinplot.
3. Swarmplot.
4. Boxenplot(Letter-Value Plot)

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
sns.boxplot(df, y='math score')
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(df, x='parental level of education', y='math score', hue='test preparation course')
plt.tight_layout()
plt.legend(loc=(1.01,0))
```

```python
plt.figure(figsize=(10,5))
sns.boxenplot(df, x='parental level of education', y='math score', hue='test preparation course')
plt.tight_layout()
plt.legend(loc=(1.01,0))
```
