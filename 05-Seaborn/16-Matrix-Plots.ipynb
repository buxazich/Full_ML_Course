# Matrix Plots - матричные графики. 
* Это визуальный аналог сводных таблиц. 
* Он показывает все данные, визуализируя числовые значения в датафрейме. 
В seaborn 2 вида матричных графиков: 
1. heatmap() - изображает распределение значений по ячейкам с помощью цветов. 
2. clustermap() - похожа на heatmap, но сначала выполняет иерархическую кластеризацию, объединяя данные в группы.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('country_table.csv')
df.head()
```

```python
df = df.set_index('Countries')
df
```

```python
sns.heatmap(data=df.drop('Life expectancy', axis=1), linewidths=0.5, annot=True, cmap='viridis');    # можно внутри метода удалить колонку
```
