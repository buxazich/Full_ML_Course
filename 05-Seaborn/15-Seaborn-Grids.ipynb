# Seaborn grids
* Это способ построения набора графиков рядом друг с другом. В основе лежит subplots() из Matplotlib, с помощью которого автоматически создаётся набор графиков для категориальной переменной. 
* Указывать кол-во колонок и строк не нужно, вместо этого мы указываем название колонки, и автоматом создаются. 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

---
## sns.catplot()

```python
sns.catplot(df, x='gender', y='math score', kind='box', row='lunch', col='test preparation course'); # т.е. метод row сплитанет еще по другой категории по строке, а col по колонке. 
```

---
## sns.pairgrid()

```python
q = sns.PairGrid(df, hue='gender')

q = q.map_upper(sns.scatterplot)
q = q.map_lower(sns.kdeplot)
q = q.map_diag(sns.histplot)

q.add_legend()
```
