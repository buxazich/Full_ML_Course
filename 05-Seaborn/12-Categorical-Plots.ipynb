# Categorical Plots - графики по категориям. 
* Они отображают статистические метрики для каждой из категорий переменной. 
* Например, среднее значение для каждой категории, или кол-во строк. 
* Это эквивалентно вызову функции groupby()

* Основные 2 вида графиков: 
1. countplot() - подсчитывает кол-во строк по категориям.
2. barplot() - общая форма отображения любой метрики по категориям. 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

---
## countplot()

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(8,5), dpi=100)
sns.countplot(df,x='division', hue='level of education')
plt.tight_layout()
```

```python
plt.figure(figsize=(8,5), dpi=100)
sns.countplot(df,x='level of education', hue='division')
plt.tight_layout()
```

```python
df['level of education'].value_counts()
```

---
## barplot()

```python
plt.figure(figsize=(8,5))
sns.barplot(df, x='level of education', y='salary', hue='division', estimator=np.mean, errorbar='sd')
plt.tight_layout()
plt.legend(loc=(1.02, 0.7))
```
