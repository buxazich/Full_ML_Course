# Seaborn 
* Это библиотека по построению статистических графиков, которая отлично работает с Pandas DataFrames для создания часто используемых графиков. 
* Seaborn построен на основе Matplotlib, но в нём мноие графики можно создать "одной строкой кода". 
* Т.к. пункт 2, можно вызывать методы plt для изменения свойств графиков Seaborn

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

## Scatter plots - диаграммы рассеяния. 
* Показывает взаимосвязь между 2 непрерывными переменными. 
* "Непрерывная" переменная - это числовая переменная, которая может принимать любые промежуточные значения. 
* В "непрерывной" переменной между 2 значениями всегда найдется какое-то 3 значение.
* Пример: возраст, высота, зарплата, температура. 

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x='sales', y='salary', hue='work experience',palette='flare', style='level of education', legend='auto')
# plt.savefig('First_scatter_plot_sns.png')
```
