# Distribution Plots - графики распределения вероятностей. 
* DP отображают одну непрерывную переменную, и помогают визуализировать атрибуты этой переменной, например среднее значение или дисперсию. 
* Три основных типа графиков:
1. Rug Plot 
2. Histogram 
3. KDE Plot

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

---
## Rug plot

```python
plt.figure(figsize=(8,5))
sns.rugplot(df, x='salary', height=0.3)
```

--- 
## Distribution plot
### sns.displot()

```python
plt.figure(figsize=(8,5))
sns.set(style='darkgrid')
sns.displot(df, x='salary', bins=20, kde=True, rug=True);
```

### sns.histplot()

```python
plt.figure(figsize=(8,5))
sns.histplot(df, x='salary')
```

### sns.kdeplot()

```python
sns.kdeplot(df,x='salary')
```

## KDE - Kernel Densiy Estimation. 
* KDE - ядерная оценка плотности. 
* Это способ оценки фукнции распределения случайной величины. 
* Простыми словами - это способ построения непрерывного графика по некоторому набору дискретных данных 


```python
np.random.seed(42)
list_samples = np.random.randint(0,100,200)
list_samples
```

```python
list_samples = pd.DataFrame(list_samples, columns=['age'])
list_samples
```

```python
sns.kdeplot(list_samples, x='age', clip=[0,100], fill=True)    # clip - для ограничения границ. 
```

```python
sns.displot(list_samples, x='age', bins=50, kde=True)
```
