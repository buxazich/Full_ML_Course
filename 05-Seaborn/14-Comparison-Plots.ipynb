# Comparison plots - графики сравнения. 
* Графики сравнения - это 2-мерные версии графиков из предыдущей темы. 
1. jointplot() - можно взять гистограммы для каждой переменной графика scatterplot, чтобы уточнить распределение данных по каждой переменной. 
2. pairplot() - быстрый способ сравнить между собой ве числовые переменные в наборе данных. Автоматически создает гистограмму для каждой переменной, и сравнительные графики scatterplot для всех комбинаций переменных. 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

--- 
## jointplot()

```python
sns.jointplot(df, x='math score', y='reading score', kind='scatter', hue='gender');
```

```python
sns.jointplot(df, x='math score', y='reading score', kind='hex');
```

```python
sns.jointplot(df, x='math score', y='reading score', kind='kde', fill=True);
```

--- 
## pairplot()

```python
sns.pairplot(df, hue='gender', corner=True) # категориальные сразу отбрасываются

# corner=True - убирает дубликаты
```
