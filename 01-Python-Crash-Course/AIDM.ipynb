Мы начнем с определения энтропии динамической системы. 

Рассмотрим динамические системы с дискретным временем. Фазовое пространство динамической системы обозначается $ M $. Оно оснащено $\sigma$-алгеброй $\mathcal{M}$ и вероятностной мерой $\mu$, определенной на $\mathcal{M}$. В общей теории эргодичности динамика задается измеримым преобразованием $ T $ пространства $ M $ в себя, сохраняющим меру $\mu$. 

Возьмем конечное разбиение $\xi = \{C_1, C_2, \ldots, C_r\}$ пространства $ M $. Оно порождает стационарный случайный процесс теории вероятностей со значениями $1, 2, \ldots, r$, если использовать формулу:
$$w_k(x) = j, \text{ если } x \in T^{-k} C_j, -\infty < k < \infty.$$

Таким образом, каждая точка $ x $ порождает реализацию случайного процесса $ w(x) = \{\ldots, w_{-n}(x), \ldots, w_0(x), w_1(x), \ldots, w_m(x), \ldots\} $. Теорема Шеннона-МакМиллана утверждает, что существует предел:

$$h(T, \xi) = \lim_{n \to \infty} \frac{1}{n} \sum_{i_1, \ldots, i_n} \mu(T^{-1} C_{i_1} \cap \ldots \cap T^{-n} C_{i_n}) \ln \mu(T^{-1} C_{i_1} \cap \ldots \cap T^{-n} C_{i_n}).$$

Определение 1. Энтропия динамической системы $ h(T) = \sup_{\xi} h(T, \xi) $, где супремум берется по всем конечным разбиениям $\xi$.

Следующая теорема является основным инструментом для вычисления $ h(T) $. Она использует понятие порождающего разбиения.

Определение 2. Разбиение $\xi$ называется порождающим разбиением (или генератором) динамической системы $(M, \mathcal{M}, \mu, T)$, если наименьшая $\sigma$-алгебра, содержащая все $ T^n C_j, -\infty < n < \infty, 1 \leq j \leq r $, совпадает с $\mathcal{M}$.

Теорема 1. Если $\xi$ является порождающим разбиением, то $ h(T) = h(T, \xi) $.

Она основана на формуле энтропии для бернуллиевых сдвигов. Доказательство теоремы для общего случая было дано в [S1]. Оно использует неравенство для условных энтропий, которое в бернуллиевом случае является равенством. Однако в то время переход от равенства к неравенству был серьезным шагом. Обсуждения с М. С. Пинскером о общих свойствах энтропии были очень полезны.

Энтропию можно определить для динамических систем с непрерывным временем, поскольку $ h(T^t) = |t| h(T^1), -\infty < t < \infty $. 

Она обладает многими полезными свойствами. Например, если $ h(T) > 0 $, то спектр унитарного оператора, соответствующего $ T $, имеет компоненту счетного спектра Лебега.

Изложение теории энтропии динамических систем можно найти во многих монографиях и учебниках, см., например, [B], [CFS], [P], [W]. Теперь известно множество примеров динамических систем с положительной энтропией даже в классе детерминированных динамических систем. Энтропия играет важную роль в теории детерминированного хаоса, поскольку она характеризует внутреннюю неустойчивость динамики и скорость расхождения близких траекторий.

Понятие метрической энтропии динамической системы, или энтропии Колмогорова, или энтропии Колмогорова-Синая, появилось в статье Колмогорова ([K1]). Это было время, когда Колмогоров интересовался и работал над несколькими задачами из теории информации, размерности функциональных пространств и т. д. Основная теорема KAM-теории появилась в работах Колмогорова немного раньше. В 1957 году Колмогоров вел семинар по динамическим системам, на котором присутствовали такие люди, как Алексеев, Арнольд, Тихомиров, Пинскер, Мешалкин, автор настоящей статьи и другие.
Проблема метрического изоморфизма динамических систем была одной из наиболее часто обсуждаемых. В своих лекциях, сопровождавших семинар, Колмогоров представил вероятностное доказательство теоремы фон Неймана об изоморфизме динамических систем с чисто точечным спектром. Основной точкой зрения в то время было то, что динамические системы, возникающие в теории вероятностей, отличаются с метрической точки зрения от систем, порожденных ОДУ и УЧП.

## Ключевые определения и объяснения (на русском)

1. Динамическая система:

    Динамическая система с дискретным временем определяется фазовым пространством $ M $, оснащенным $\sigma$-алгеброй $\mathcal{M}$ и вероятностной мерой $\mu$. Измеримое преобразование $ T: M \to M $ сохраняет меру $\mu$, то есть $\mu(T^{-1}(A)) = \mu(A)$ для любого измеримого множества $ A \in \mathcal{M} $.

    Объяснение: Динамическая система моделирует эволюцию системы во времени. Фазовое пространство $ M $ представляет все возможные состояния, а $ T $ описывает, как состояния изменяются от одного шага времени к следующему, сохраняя вероятностную структуру.

2. Пространство Лебега:

    Пространство с мерой, изоморфное единичному интервалу $[0,1]$ с обычной мерой Лебега.
    Объяснение: Пространства Лебега стандартны в теории эргодичности, так как обеспечивают удобный каркас для изучения сохраняющих меру преобразований.

3. Конечное разбиение:

    Конечное разбиение $\xi = \{C_1, C_2, \ldots, C_r\}$ пространства $ M $ делит фазовое пространство на конечное число измеримых подмножеств.
    Объяснение: Разбиения используются для дискретизации фазового пространства, что позволяет изучать поведение системы, наблюдая, в какой элемент разбиения попадает точка $ x \in M $ при итерациях $ T $.

4. Стационарный случайный процесс:

    Для разбиения $\xi$ преобразование $ T $ порождает случайный процесс, где $ w_k(x) = j $, если $ x \in T^{-k} C_j $, присваивая индекс $ j $ (от 1 до $ r $) каждой итерации точки $ x $.
    Объяснение: Этот процесс связывает последовательность символов с каждой траекторией, превращая динамическую систему в вероятностный объект, анализируемый методами теории вероятностей.

5. Теорема Шеннона-МакМиллана:

    Для разбиения $\xi$ энтропия $ h(T, \xi) $ определяется как:
    $$h(T, \xi) = \lim_{n \to \infty} \frac{1}{n} \sum_{i_1, \ldots, i_n} \mu(T^{-1} C_{i_1} \cap \ldots \cap T^{-n} C_{i_n}) \ln \mu(T^{-1} C_{i_1} \cap \ldots \cap T^{-n} C_{i_n}).$$

    Объяснение: Эта теорема количественно определяет среднюю неопределенность (или информационное содержание) на шаг времени в случайном процессе, вызванном $ T $ и $\xi$.

6. Метрическая энтропия динамической системы:

    Энтропия $ h(T) = \sup_{\xi} h(T, \xi) $, где супремум берется по всем конечным разбиениям $\xi$.
    Объяснение: Метрическая энтропия (или энтропия Колмогорова-Синая) измеряет сложность или случайность системы, будучи метрическим инвариантом.

7. Порождающее разбиение:

    Разбиение $\xi = \{C_1, \ldots, C_r\}$ является порождающим, если наименьшая $\sigma$-алгебра, содержащая все множества $ T^n C_j $, для всех $ n $ и $ j $, совпадает с $\mathcal{M}$.
    Объяснение: Порождающее разбиение полностью описывает динамику системы, упрощая вычисление энтропии.

8. K-система (Колмогоровская система):

    Динамическая система является K-системой, если она обладает свойствами, схожими с квазирегулярными системами, напоминающими регулярные стационарные процессы.
    Объяснение: K-системы обладают высокой степенью перемешивания и положительной энтропией, что указывает на сильное хаотическое поведение.

Заключение

Метрическая энтропия, разработанная Колмогоровым и расширенная Синаем, является фундаментальным инвариантом в теории эргодичности, измеряющим сложность и непредсказуемость динамических систем. Документ описывает ее определение, методы вычисления (через порождающие разбиения) и исторический контекст, подчеркивая ее роль в различении хаотического и детерминированного поведения. Ключевые понятия, такие как порождающие разбиения и теорема Шеннона-МакМиллана, упрощают вычисления энтропии, а исторические анекдоты подчеркивают коллаборативный характер ее разработки. Для дальнейшего изучения указанные источники предоставляют более глубокое погружение в теорию эргодичности и применения энтропии.
