```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/cancer_tumor_data_features.csv')
df.head()
```

```python
from sklearn.datasets import load_breast_cancer

breast = load_breast_cancer()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_X = scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA


pca_model = PCA(n_components=2)
pc_result = pca_model.fit_transform(scaled_X)

plt.scatter(x=pc_result[:,0], y=pc_result[:,1],c=breast['target']);
```

```python
pca_model.components_
```

```python
df_components = pd.DataFrame(pca_model.components_, index=['PC1','PC2'], columns=df.columns)
df_components
```

```python
plt.figure(figsize=(20,3), dpi=150)
sns.heatmap(data=df_components, annot=True)
```

```python
pca_model.explained_variance_ratio_
```

```python
np.sum(pca_model.explained_variance_ratio_)
```

```python
pca_30 = PCA(n_components=30)
pca_30.fit(scaled_X)
pca_30.explained_variance_ratio_
```

```python
np.sum(pca_30.explained_variance_ratio_)
```

```python
explained_var = []

for n in range(1,30):
    pca = PCA(n_components=n)
    pca.fit(scaled_X)
    explained_var.append(np.sum(pca.explained_variance_ratio_))
    
plt.plot(range(1,30), explained_var, '--o')
plt.xlabel('ÐšÐ¾Ð»-Ð²Ð¾ Ð³Ð»Ð°Ð²Ð½Ñ‹Ñ… ÐºÐ¾Ð¼Ð¿Ð¾Ð½ÐµÐ½Ñ‚')
plt.ylabel('Variance explained')
```
