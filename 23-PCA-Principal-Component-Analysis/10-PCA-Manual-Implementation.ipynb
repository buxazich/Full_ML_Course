```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/cancer_tumor_data_features.csv')
df.head()
```

```python
sns.heatmap(df);
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_X = scaler.fit_transform(df)

round(scaled_X.mean(),2)
```

```python
scaled_X.mean(axis=0)
```

```python
covariance_matrix = np.cov(scaled_X, rowvar=False) #rowvar - ÐºÐ¾Ð²Ð°Ñ€Ð¸Ñ Ð¼ÐµÐ¶Ð´Ñƒ Ð¾Ð±ÑŠÐµÐºÑ‚Ð°Ð¼Ð¸ (Ð½ÑƒÐ¶Ð½Ñ‹ Ñ‚Ð¾Ð»ÑŒÐºÐ¾ Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ¸)
covariance_matrix
```

```python
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
eigen_values
```

```python
eigen_vectors.shape
```

```python
num_components = 2

sorted_key = np.argsort(eigen_values)[::-1][:num_components]
sorted_key
```

```python
eigen_values, eigen_vectors = eigen_values[sorted_key], eigen_vectors[:,sorted_key]
```

```python
principal_components = np.dot(scaled_X, eigen_vectors)
principal_components #ÐºÐ¾Ð¾Ñ€Ð´Ð¸Ð½Ð°Ñ‚Ñ‹ Ð² Ð½Ð¾Ð²Ð¾Ð¼ Ð¿Ñ€Ð¾ÑÑ‚. Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ¾Ð²
```

```python
plt.scatter(x=principal_components[:,0], y=principal_components[:,1]);
```
